```python
# Q9
import pandas as pd

data = {
 'Movies': ['KPop Demon Hunters', 'The Pursuit of Happyness', 'Cars 2', 'Interstellar', 'Spider-Man', 'Superman', 'Suicide Squad', 'Oppenheimer', 'The Minecraft Movie', 'Big Fish'],
 'Likes': ['I loved the soundtrack.', 'It is a good motivational story.', 'It\'s a departure from the genre of Cars.', 'I like that it\'s about space.', 'I like the green goblin', 'The story was really good.', 'Margot Robbie killed it as Harley Quinn', 'The history was really well researched, and the special effects were great.', 'It was fun and lived up to the hype.', 'It\'s a great sleep aid.'],
 'Dislikes': ['The title is stupid.', 'I hate the way they spell happyness in the movie title', 'It\'s a departure from the genre of Cars.', 'It\'s too sad.', 'I dislike the character of Mary Jane.', 'Krypto could have benefited from less screen time.', 'The plot was really messy.', 'It was way too long.', 'It didn\'t follow the game.', 'Too slow of a pace.'],
 'Rating': [5.0 ,4.5, 4.9, 5.0, 3.5, 4.0, 4.1, 3.9, 2.5, 1.2]
}
df = pd.DataFrame(data)

print(df.columns.tolist())
df
```

```python
# Q10
import math

modified_df = df.copy()

modified_df['Movies'] = df['Movies'].str.replace(' ', '')
modified_df['Likes'] = df['Likes'].str.replace(' ', '')
modified_df['Dislikes'] = df['Dislikes'].str.lower()
modified_df['Rating'] = df['Rating'].apply(lambda x: math.floor(x))

modified_df
```

```python
# Q11
import matplotlib.pyplot as plt

plt.hist(df['Rating'])

plt.show()
print(df['Rating'].mean())
print(df['Rating'].std())
```

```python
# Q12
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.scatter(df['Movies'], df['Rating'])
ax1.set_xticks(range(len(df)))
ax1.set_xticklabels([])

ax2.bar(df['Movies'], df['Rating'])
ax2.set_xticks(range(len(df)))
ax2.set_xticklabels(df['Movies'], rotation=45, ha='right')

plt.tight_layout()
plt.show()
```

```python
# Q13
df['ZRatings'] = df['Rating'].apply(lambda x: (x - df['Rating'].mean()) / df['Rating'].std())

plt.hist(df['ZRatings'], bins=10)
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized Ratings')

plt.show()
```

```python
# Q14
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(df['Rating'], bins=10)
ax1.set_title('Original Ratings')
ax1.set_xlabel('Rating')
ax1.set_ylabel('Frequency')

ax2.hist(df['ZRatings'], bins=10)
ax2.set_title('Z-Scores')
ax2.set_xlabel('Z-Score')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Q15
import numpy as np

df['RandomRatings'] = df['Rating'].apply(lambda x: np.random.uniform(1, 100))

# Q15a
plt.hist(df['RandomRatings'])

plt.show()
print(df['RandomRatings'].mean())
print(df['RandomRatings'].std())

# Q15b
fig, (ax1, ax2) = plt.subplots(2, 1)
ax1.scatter(df['Movies'], df['RandomRatings'])
ax1.set_xticks(range(len(df)))
ax1.set_xticklabels([])

ax2.bar(df['Movies'], df['RandomRatings'])
ax2.set_xticks(range(len(df)))
ax2.set_xticklabels(df['Movies'], rotation=45, ha='right')

plt.tight_layout()
plt.show()

# Q15c
df['ZRandomRatings'] = df['RandomRatings'].apply(lambda x: (x - df['RandomRatings'].mean()) / df['RandomRatings'].std())

plt.hist(df['ZRandomRatings'], bins=10)
plt.xlabel('Z-Score')
plt.ylabel('Frequency')
plt.title('Histogram of Normalized Ratings')

plt.show()

# Q15d
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.hist(df['RandomRatings'], bins=10)
ax1.set_title('Original Ratings')
ax1.set_xlabel('Rating')
ax1.set_ylabel('Frequency')

ax2.hist(df['ZRandomRatings'], bins=10)
ax2.set_title('Z-Scores')
ax2.set_xlabel('Z-Score')
ax2.set_ylabel('Frequency')

plt.tight_layout()
plt.show()
```
