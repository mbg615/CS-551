```python
# Q9
mean = df['employee_salary'].mean()
std = df['employee_salary'].std()

# Define the threshold for outliers (3 standard deviations)
lower_bound = mean - 3 * std
upper_bound = mean + 3 * std

# Identify outliers
outliers = df[(df['employee_salary'] < lower_bound) | (df['employee_salary'] > upper_bound)]

# Remove outliers
df = df[(df['employee_salary'] >= lower_bound) & (df['employee_salary'] <= upper_bound)]
df
```

```python
# Q10
from sklearn.datasets import load_iris
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt

iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df['species'] = pd.Categorical.from_codes(iris.target, iris.target_names)

versicolor = df[df['species'] == 'versicolor']
model = smf.ols('Q("petal width (cm)") ~ Q("petal length (cm)")', data=versicolor).fit()

plt.scatter(versicolor['petal length (cm)'], versicolor['petal width (cm)'], color='blue', label='Data points')
plt.plot(versicolor['petal length (cm)'], model.predict(versicolor), color='red', label='Regression line')
plt.xlabel('Petal Length (cm)')
plt.ylabel('Petal Width (cm)')
plt.title('Petal Length vs Petal Width (Versicolor)')
plt.legend()
plt.show()
```
